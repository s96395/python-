**Ex3**
 : 撰寫一python程式，輸入一正整數N，列出所有小於N的質數

```python
N = eval(input('請輸入一個正整數:'))
# N = int(input('請輸入一個正整數:'))
def isPrime(n):
  #判斷n是否為質數
  count = 0
  for i in range(1,n+1):
    if n%i==0:
      count+=1

  if count==2:
    return True
  else:
    return False

for i in range(2,N+1):
  if isPrime(i):
    print(i)
```
