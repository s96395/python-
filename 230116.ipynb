計算薪水

```python
#######variables
hourly_salary = 125
hour = 8
days = 300
monthly_fee = 9000
months = 12

######main(process/algorithm...)
annual_salary = hourly_salary * hour * days
annual_fee = monthly_fee * months
annual_savings = annual_salary - annual_fee


print(annual_savings)
```

```python
#######variables
hourly_salary = 125  #時薪
hour = 8  #工作時數
days = 300  #天數
monthly_fee = 9000  #月花費
months = 12  #月份

######main(process/algorithm...)
annual_salary = hourly_salary * hour * days  #年收入
annual_fee = monthly_fee * months  #年支出
annual_savings = annual_salary - annual_fee  #年存款


print(annual_savings)
```

稅收

```python
print("年度   所得稅   營業稅    證交稅")
print("%4s   %6.2f   %6.2f    %6.2f"  %  ("2017", 98.34, 90.2, 104.79))
print("%4s   %6.2f   %6.2f    %6.2f"  %  ("2016", 83, 110.5, 82.45))
print("%4s   %6.2f   %6.2f    %6.2f"  %  ("2015", 98, 79.32, 102))
```

```python
print("年度   所得稅   營業稅    證交稅")
print("{:4s}   {:6.2f}   {:6.2f}    {:6.2f}".format("2017", 98.34, 90.2, 104.79))
print("{:4s}   {:6.2f}   {:6.2f}    {:6.2f}".format("2016", 83, 110.5, 82.45))
print("{:4s}   {:6.2f}   {:6.2f}    {:6.2f}".format("2015", 98, 79.32, 102))
```
