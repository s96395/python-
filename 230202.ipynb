CH3 p.44

```python
fruits = ["香蕉","蘋果","橘子","西瓜"]
while True:
  print("串列元素有:",fruits)
  fruit = input("請輸入要刪除的水果(直接輸入Enter結束):")
  if(fruit==""):
    break
  n=fruits.count(fruit)
  if(n>0): #串列元素存在
    fruits.remove(fruit)
  else:
    print(fruit,"不在串列中!")

print('final')
```
