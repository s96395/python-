```python
import pandas as pd
from pandas import Series, DataFrame
```

```python
attri_data1 = {'ID':['100','101','102','103','104'],
               'City':['Tokyo','Osaka','Kyoto','Hokkaido','Tokyo'],
               'Birth_year':[1990,1989,1992,1997,1982],
               'Name':['Hiroshi','Akiko','Yuki','Satoru','Steve']}

attri_data_frame1 = DataFrame(attri_data1)

print(attri_data_frame1)
```

```python
attri_data_frame_index1 = DataFrame(attri_data1,index=['a','b','c','d','e'])
print(attri_data_frame_index1)
```

```python
# 取出特定某一行
attri_data_frame_index1.Birth_year
```

```python
# 取出特定多行
attri_data_frame_index1[['ID', 'Birth_year']]
```

```python
attri_data_frame_index1.drop(['Birth_year'], axis = 1)
```

```python
#只取出City為Tokyo的資料
attri_data_frame_index1[attri_data_frame_index1['City']=='Tokyo']
```

```python
#單獨取出條件，可觀察是True或False
attri_data_frame_index1['City']=='Tokyo'
```

```python
#想指定多個條件時，可適用isin（串列）
attri_data_frame_index1[attri_data_frame_index1['City'].isin(['Tokyo','Osaka'])]
```

```python
attri_data2 = {'ID':['100','101','102','105','107'],
               'Math':[50,43,33,76,98],
               'English':[90,30,20,50,30],
               'Sex':['M','F','F','M','M']}
attri_data_frame2 = DataFrame(attri_data2)
attri_data_frame2
```

```python
pd.merge(attri_data_frame_index1,attri_data_frame2)
```

```python
attri_data_frame2.groupby('Sex')['Math'].mean()
```

```python
attri_data1 = {'ID':['1','2','3','4','5'],
               'Sex':['F','F','M','M','F'],
               'Money':[1000,2000,500,300,700],
               'Name':['Saito','Horie','Kondo','Kawada','Matsubara']}

attri_data_frame1 = DataFrame(attri_data1)
```

```python
#找出Money為500以上的人
attri_data_frame1['Money']>=500
```

```python
#請分別計算男女(MF分開看)的平均Money
attri_data_frame1.groupby('Sex')['Money'].mean()#平均值
```
