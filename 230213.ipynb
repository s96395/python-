```python
import random as r
#function
def lo_system(low,upper,n):
  list1=r.sample(range(low,upper+1),n) #1 2 3 50-1
  special=list1.pop()
  list1.sort()
  return list1,special

def print_result(list1,special=None):
  print('本期中獎號碼為:',end='')
  for i in range(len(list1)):
    print(str(list1[i]),end=' ')
  if special == None:
    print('本期特別號為:'+str(special))

if __name__ == '__main__':
  low = 1
  upper = 49
  n = 7

  list1,special = lo_system(low,upper,n)
  print_result(list1,special)
```

```python
def reverse(n): #22322
  reverseN = 0
  while n != 0:
    r = n % 10
    #print(r,end='') # 2 2 3 2 2 
    reverseN = reverseN * 10 + r #22322
    n//=10 #5
  return reverseN #22322

def isPalindrome(n): #22322
  return n == reverse(n) #22322==22322
```

```python
x = eval(input('請輸入1個數值=')) #22322
if isPalindrome(x):
  print('這是回文數')
else:
  print('這不是回文數')
```

```python
def fib(n): #4
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fib(n-1)+fib(n-2) #return 2+1

print(fib(4))
```

```python
print('下列是前4個Fibonacci數列')
for index in range(4):
  print('%4d' % fib(index),end='') #index=4
```
