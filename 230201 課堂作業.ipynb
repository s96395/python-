CH2實作題3 p.53

```python
i=input("溫度轉換選擇\n1:華氏溫度轉成攝氏溫度\n2:攝氏溫度轉成華氏溫度\n=")
if i == "1": #1:華氏溫度轉成攝氏溫度
  f = input("請輸入華氏溫度:")
  c = (int(f)-32) * 5 / 9
  print("華氏 %s 等於攝氏 %4.1f" % (f,c))
elif i == "2": #2:攝氏溫度轉成華氏溫度
  c = input("請輸入攝氏溫度:")
  f = int(c) * 9 / 5 + 32
  print("攝氏 %s 等於華氏 %4.1f" % (c,f))
else:
  print("輸入錯誤")
```

CH3實作題4 p.30

```python
p=1
n = int(input("請輸入正整數:")) #7

counter=0 #正因數的個數

print(n,"的因數有",end=" ")
while(p<=n):
  if(n%p==0):
    print(p,end=" ")
    counter+=1
  p+=1

print()
if(counter==2):
  print(n,"是質數")
else:
  print(n,"不是質數")
```

CH3實作題7 p.33

```python
male = 5 #公雞
female = 3 #母雞
child = 1/3 #小雞

budget = 100 #成本
number = 100 #隻數

for i in range(100): #male
  for j in range(100): #female
    for k in range(100): #child
      if(i+j+k == number) and (i*male+j*female+k*child == 100):
        print(i,j,k)
```

```python
male = 5 #公雞
female = 3 #母雞
child = 1/3 #小雞

budget = 100 #成本
number = 100 #隻數

import time
start = time.perf_counter()
count = 0

for i in range(int(budget/male)+1+1): #100/5=20
  for j in range(int(budget/female)+1+1): #100/3=33.3/33+1 34
    count += 1
    k = number - i - j #i+j+k=100
    if(i*male+j*female+k*child == budget):
        print(i,j,k)
end=time.perf_counter()
print('count:',count)
print('time:',end-start)
```
