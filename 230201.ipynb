p.15 ★用蒙地卡羅方法算圓周率




```python
import random as r
points = 100000
count = 0

for i in range(points):
    x, y = r.random(), r.random()
    if ((x**2+y**2)**0.5) <= 1:
        count += 1

print(4*(count/points))
```

```python
total = 0
person = 0
score = 0
while(score!=-1):
  person += 1
  total += score
  score = int(input(f"請輸入第{person}位學生的成績:"))
average = total/(person-1)
print(f"本班總成績: %d 分,平均成績: %5.2f 分" % (total,average))
```
