```python
import string
#abc = string.ascii_lowercase
abc = 'abcdefghijklmnopqrstuvwxyz'
encry_dict = {} #dictionary
front3 = abc[:3] #abc
end23 = abc[3:] #d~z
subText = end23 + front3
ency_dict = dict(zip(subText,abc)) #建立字典
print("列印編碼字典\n",encry_dict) #列印字典

msgTest = 'python' #測試字串
cipher=[]
for i in msgTest: #執行每個字元加密
  v=encry_dict[i] #加密
  cipher.append(v) #加密結果
ciphertext=''.join(cipher) #將串列轉成字串

print('原始字串',msgTest)
print('加密字串',ciphertext)

results=[]
for c in ciphertext:
  for k,v in encry_dict.items():
    if v==c:
      results.append(k)
print('解密:')
print(''.join(results))
```

```python
def isPrime(num): #7
  """測試num是否質數"""
  for n in range(2,num): #2,3,4,...,(7-1)
    if num % n == 0: #餘數的計算
      return False #Flag=False表示num不是質數
  return True #True是質數 False不是質數

if isPrime(66987114):
  print('是質數')
else:
  print('不是質數')
```

```python
def calsum(*params): #a tuple named params
  print(params,type(params))
  total = 0
  for param in params: #tuple iteration
    total += param
  return total

print('不定數目範例:')
print('2個參數:calsum(4,5)=%d'% calsum(4,5))
print('3個參數:calsum(4,5,12)=%d'% calsum(4,5,12))
print('4個參數:calsum(4,5,12,8)=%d'% calsum(4,5,12,8))
```
