```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.taiwanlottery.com.tw'
r = requests.get(url)
print(r.status_code)
```

```python
sp=BeautifulSoup(r.text,'lxml')
#找到威力彩的區塊
#開獎區於class=contents_box02的<div>中
datas = sp.find('div',class_='contents_box02')
```

```python
from pprint import pprint
pprint(datas.text)
```

```python
#開獎期數
#該期的期號在class_=font_black15的<span>中
title = datas.find('span',class_='font_black15').text
print('威力彩期數:',title)
#開獎號碼
#所有開獎號碼在class_=ball_tx ball_green的<div>中
nums=datas.find_all('div',class_='ball_tx ball_green')
#nums
#開出順序
print('開出順序:',end=' ')
for i in range(0,6):
  print(nums[i].text,end=' ')
print()
#大小順序
print('大小順序:',end=' ')
for i in range(6,len(nums)):
  print(nums[i].text,end=' ')
#第二區
#第二區的開獎號碼在class=ball_red的<div>中
num = datas.find('div',class_='ball_red').text
print('\n第二區',num)
```

```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt

month,high,low = [],[],[] #empty lists
#requrests data
url = 'https://www.twse.com.tw/exchangeReport/FMSRFK?'
#headers
params = {
   'response':'html',
   'date':'20200101',
   'stockNo':'2330'
    }
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'
}
content = requests.get(url, headers=headers, params=params).text
```

```python
#html analysis
sp = BeautifulSoup(content, "lxml")
datas = sp.select("table")[0] #the whole table
title=datas.find("div").text.replace(" ","") #title
rows = datas.select("tbody tr") #every month
```

```python
for row in rows:
    cols = row.select("td")
    month.append(int(cols[1].text))  #月份
    high.append(float(cols[2].text)) #最高價
    low.append(float(cols[3].text))  #最低價
```

```python
#plot
plt.plot(month, high, linewidth=2.0,label="最高價")
plt.plot(month, low, linewidth=2.0,label="最低價")
plt.legend()
plt.title(title) #圖表標題
plt.xlabel("月份")
plt.ylabel("金額")
plt.rcParams["font.sans-serif"] = "Microsoft JhengHei" #也可設mingliu或DFKai-SB
plt.show()
```
