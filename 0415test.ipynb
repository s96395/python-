使用 split 函式，分割字串

```python
test = 'hello_world_yoyo'
print(test.split("_"))
```

使用 join 函式將資料轉換成字串

```python
test = ['hello','world','yoyo']
print('_'.join(test))
```

透過 for 迴圈打印出列表中的資料

```python
test = ['hello','world','yoyo']
#定義一個空字串
j =''
#透過 for 迴圈打印出列表中的資料
for i in test:
  j = j+'_'+i
#因為透過上面的字串拼接，得到的資料是“_hello_world_yoyo”，前面會多一個下劃線_，所以把這個下劃線去掉
print(j.lstrip('_'))
```

把字串 s 中的每個空格替換成”%20”。

使用 replace 函式，替換字元

```python
s = 'we are happy.'
print(s.replace(' ','%20'))
```

for 迴圈列印 99 乘法表

```python
for i in range(1, 10):
    for j in range(1, i+1):
        print('{}x{}={}\t'.format(j, i, i*j), end='')
    print()
```

while 迴圈實現 99 乘法表

```python
i = 1
while i <= 9:
  j = 1
  while j <= i:
    print('%d*%d=%-2d'%(i,j,i*j),end=' ')
    # %d：整数的占位符，'-2'代表靠左对齐，两个占位符
    j += 1
  print()
  i += 1
```
