時間序列資料處理

```python
#下載並安裝pandas-datareader函式庫。

import pandas_datareader.data as pdr
```

```python
#使用2001/1/2至2016/12/30為止的美元日幣匯率資料（DEXJPUS）。
#這是美日的匯率資料，因此有遺漏的日程（假日等）。

!pip install pandas_datareader
start_date = '2001/1/2'
end_date = '2016/12/30'

fx_jpusdata = pdr.DataReader('DEXJPUS','fred',start_date,end_date)
```

```python
#讀取的fx_jpusdata前方取出5列。

fx_jpusdata.head()
```

```python
#觀察2016年4月的資料。
fx_jpusdata['2016-04']
```

```python
#只取出月底的匯率。

fx_jpusdata.resample('M').last().head()
```

```python
#遺漏2001/1/6的資料

fx_jpusdata.resample('D').last().head()
```

```python
#以前一天的值來填補處理。

fx_jpusdata.resample('D').ffill().head()
```

```python
#將資料往後偏移1個。

fx_jpusdata.shift(1).head()
```

```python
#如此，即可計算出前一天匯率與當日匯率的比例

fx_jpusdata_ratio = fx_jpusdata / fx_jpusdata.shift(1)
fx_jpusdata_ratio.head()
```

```python
#請使用讀取的fx_jpusdata資料，製作每年的平均值之推移資料。
fx_jpusdata.resample('Y').mean().head()
```

```python
#製作3天的移動平均線。
fx_jpusdata.head()
```

```python
#使用rolling方法，計算3天的移動平均之結果。

fx_jpusdata.rolling(3).mean().head()
```

```python
#製作20天的移動平均資料。但請刪除NaN。其中如果有不存在紀錄的資料，不需要特地填補。

fx_jpusdata.rolling(20).mean().head(30)
```

```python
#將NaN刪除(dropna)

fx_jpusdata.rolling(20).mean().dropna().head(30)
```
